```python
import urllib.parse, http.client, json

headers = {
    'Ocp-Apim-Subscription-Key':subscription_key
}

url = 'https://www.arukikata.co.jp/com_img/city/large/p_PAR.jpg'
body = {
    'url':url
}

params = urllib.parse.urlencode({
    'visualFeatures':'description,color',
    'details':'landmarks',
    'language':'en'
})

try:
    conn = http.client.HTTPSConnection('westus.api.cognitive.microsoft.com')
    conn.request('POST', '/vision/v3.0/analyze?%s' % params, str(body), headers)
    response = conn.getresponse()
    jsonData = response.read()
    celebrities_data = json.loads(jsonData)
    print(json.dumps(celebrities_data, sort_keys=True, indent=2)) # jsonをわかりやすく表示する
    conn.close()
except Exception as ex:
    print(ex)
```

# Describe Image

```python
# maxCandidatesで出てくる候補の数を指定
describe_params = urllib.parse.urlencode({
    'maxCandidates':'3',
    'language':'en',
})

try:
    conn = http.client.HTTPSConnection('westus.api.cognitive.microsoft.com')
    conn.request('POST', '/vision/v3.0/describe?%s' % describe_params, str(body), headers)
    response = conn.getresponse()
    jsonData = response.read()
    data = json.loads(jsonData)
    print(json.dumps(data, sort_keys=True, indent=2))
    conn.close()
except Exception as ex:
    print(ex)
```

# OCR

```python
ocr_url = 'https://images-na.ssl-images-amazon.com/images/I/61tAWrb9HQL.jpg'

# detectOrientationをtrueにすると、いろんな方向を検知可能
ocr_params = urllib.parse.urlencode({
    'language':'en',
    'detectOrientation':False
})

ocr_body = {
    'url':ocr_url
}

try:
    conn = http.client.HTTPSConnection('westus.api.cognitive.microsoft.com')
    conn.request('POST', '/vision/v3.0/ocr?%s' % describe_params, str(ocr_body), headers)
    response = conn.getresponse()
    jsonData = response.read()
    data = json.loads(jsonData)
    print(json.dumps(data, sort_keys=True, indent=2))
    conn.close()
except Exception as ex:
    print(ex)
```

```python
import requests
from io import BytesIO
from PIL import Image, ImageDraw
import matplotlib.pyplot as plt

# 画像を取得
image_response = requests.get(ocr_url)
image = Image.open(BytesIO(image_response.content))

drawing = ImageDraw.Draw(image)

plt.imshow(image)
plt.show()
```

```python
region_color = 'blue'
line_color = 'green'
word_color = 'red'

# Regionのループ
for region in data['regions']:
    # 座標を取得
    region_box = region['boundingBox'] # region の left top width heightが得られる
    values = region_box.split(',')
    print("region:",values)
    left = int(values[0])
    top = int(values[1])
    width = int(values[2])
    height = int(values[3])
    
    # 描画
    drawing.line([(left,top), (left+width, top)], fill=region_color, width=7)
    drawing.line([(left+width,top), (left+width, top+height)], fill=region_color, width=7)
    drawing.line([(left,top+height), (left+width, top+height)], fill=region_color, width=7)
    drawing.line([(left,top), (left, top+height)], fill=region_color, width=7)
    
    # lineのループ
    for line in region['lines']:
        line_box = line['boundingBox']
        values = line_box.split(',')
        left = int(values[0])
        top = int(values[1])
        width = int(values[2])
        height = int(values[3])
        drawing.line([(left,top), (left+width, top)], fill=line_color, width=5)
        drawing.line([(left+width,top), (left+width, top+height)], fill=line_color, width=5)
        drawing.line([(left,top+height), (left+width, top+height)], fill=line_color, width=5)
        drawing.line([(left,top), (left, top+height)], fill=line_color, width=5)
  
        # wordのループ
        for word in line['words']:
            word_box = word['boundingBox']
            text = word['text']
            print(text)
            values = word_box.split(',')
            left = int(values[0])
            top = int(values[1])
            width = int(values[2])
            height = int(values[3])
            drawing.line([(left,top), (left+width, top)], fill=word_color, width=3)
            drawing.line([(left+width,top), (left+width, top+height)], fill=word_color, width=3)
            drawing.line([(left,top+height), (left+width, top+height)], fill=word_color, width=3)
            drawing.line([(left,top), (left, top+height)], fill=word_color, width=3)

image.save('ocr.jpg')

plt.figure(figsize=(15,8))
plt.imshow(image)
plt.show()
```

# Handwritting recognition

* https://westus.dev.cognitive.microsoft.com/docs/services/5adf991815e1060e6355ad44/operations/587f2c6a154055056008f200

```python
headers = {
    'Content-Type':'application/octet-stream', # urlを使わず、ローカル画像を使うため
    'Ocp-Apim-Subscription-Key':subscription_key,
    'Accept':'application/json'
}

params = urllib.parse.urlencode({
    'mode':'Handwritten'
})

with open('Test1.png', 'rb') as body:
    conn = http.client.HTTPSConnection('westus.api.cognitive.microsoft.com')
    conn.request('POST', '/vision/v2.0/recognizeText?%s' % params, body, headers)
    response = conn.getresponse()
    data = response.read()
    print(data) # すぐに結果が返ってこないので空白になる
    operation_location = response.getheader('Operation-Location')
    print(operation_location) # ここで分析が実行されている
    conn.close()
```

```python
# 時間がたつと結果を取得できる
response = requests.get(operation_location, headers=headers) # headerを指定してgetメソッドで結果を取得可能
jsonData = response.content
data = json.loads(jsonData)
print(json.dumps(data, sort_keys=True, indent=2))
```
